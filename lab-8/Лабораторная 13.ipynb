```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def predict(x):
    dt = 0.01
    
    for i in range(N-1):
        x.append(
            x[-1] * np.exp((mu - sigma ** 2 / 2) * dt + sigma * np.random.normal(0, np.sqrt(dt)))
        )
        
    return x
```

```python
mu = 0.8
sigma = 0.3
N = 31
t = np.linspace(0, N, N)

dollar = predict([71.1])
euro = predict([73.07])

plt.figure(figsize=(15, 10))
plt.plot(t, dollar, label='dollar')
plt.plot(t, euro, label='euro')
plt.xlabel(f'Dollar max = {max(dollar):.2f}, min = {min(dollar):.2f}\nEuro max = {max(euro):.2f}, min = {min(euro):.2f}')
plt.legend()
plt.show()
```
